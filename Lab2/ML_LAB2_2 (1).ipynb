```python
# Identify unique categories in 'ACFT_DMG_DESC'
categories = df2['ACFT_DMG_DESC'].unique().tolist()

print("Unique categories in 'ACFT_DMG_DESC':")
print(categories)
```

```python
# Create new columns for each category (except the first one to avoid multicollinearity)
for category in categories[1:]:
    col_name = f'ACFT_DMG_DESC_{category}'
    df2[col_name] = 0 # Initialize with zeros

# Populate new columns based on the original 'ACFT_DMG_DESC' column
for index, row in df.iterrows():
    original_category = row['ACFT_DMG_DESC']
    if original_category in categories[1:]:
        col_name = f'ACFT_DMG_DESC_{original_category}'
        df2.at[index, col_name] = 1

# Drop the original 'ACFT_DMG_DESC' column
df2.drop('ACFT_DMG_DESC', axis=1, inplace=True)

df2.columns
```
