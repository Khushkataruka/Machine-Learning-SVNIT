```python
import numpy as np
import random
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def dice_simulation(num_rolls, num_experiments):
    rolls = np.random.randint(1, 7, size=(num_experiments, num_rolls))  # matrix of experiments
    sums = np.sum(rolls, axis=1)  # sum across each experiment
    return sums
```

```python
dice_rolls_list = [1, 2, 3, 5, 10, 20, 50]

experiment_sizes = [100, 500, 1000, 5000, 10000, 50000, 100000]
```

```python
results_summary = []

for num_rolls in dice_rolls_list:
    for num_experiments in experiment_sizes:
        results = dice_simulation(num_rolls, num_experiments)

        # Empirical mean and variance
        mean_emp = np.mean(results)
        var_emp = np.var(results, ddof=0)

        # Theoretical mean and variance
        mean_theo = num_rolls * 3.5
        var_theo = num_rolls * 35 / 12

        results_summary.append((num_rolls, num_experiments, mean_emp, mean_theo, var_emp, var_theo))

        # Plot histogram
        plt.figure(figsize=(6,4))
        plt.hist(results, bins=30, density=True, alpha=0.7, color='lightgreen', edgecolor='black')
        plt.title(f"Dice={num_rolls}, Experiments={num_experiments}\nMean={mean_emp:.2f}, Var={var_emp:.2f}")
        plt.xlabel("Sum of dice rolls")
        plt.ylabel("Probability Density")
```

```python
summary_df = pd.DataFrame(results_summary, columns=["Dice Rolls", "Experiments",
                                                    "Empirical Mean", "Theoretical Mean",
                                                    "Empirical Variance", "Theoretical Variance"])
summary_df.head(49)
```
