```python
import pandas as pd
import numpy as np
df=pd.read_csv('advertising.csv')
df.head()
```

```python
# @title Newspaper vs Sales

from matplotlib import pyplot as plt
df.plot(kind='scatter', x='Newspaper', y='Sales', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
x=df.drop('Sales',axis=1)
y=df['Sales']
```

```python
#x and y must be numpy array
def Linear_Regression(x,y,learning_rate):
  weights=np.zeros((1,x.shape[1]))
  bias=0
  learning_rate=learning_rate
  epochs=1000
  for i in range(epochs):
    y_pred=np.dot(weights,x.T)+bias #(1,200
    print(y.shape,y_pred.shape)
    loss=np.mean((y-y_pred)**2)
    dw=2*(y_pred-y).dot(x)
    db=2*(y_pred-y)
    weights=weights-learning_rate*dw
    bias=bias-learning_rate*db
    print(f"Loss:{loss},Epoch:{i}\n{'-'*50}\n")
  return weights,bias
```

```python
y=np.array(y)
y=y.T
x=np.array(x)
w,b=Linear_Regression(x,y,1e-7)
print(w.shape,b.shape)
```

```python
def predict(x,w,b):
  x=np.array(x)
  y_pred=np.dot(w,x.T)+b
  return y_pred
```

```python
b=np.mean(b)
```

```python
predict([238,37,70],w,b)
```

```python
# Radio vs Sales
df.plot(kind='scatter', x='Radio', y='Sales', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
plt.title('Radio vs Sales')
plt.show()
```

**Interpretation of the plots and the trained model:**

From the scatter plots, we can observe the following:

*   **Newspaper vs Sales:** The scatter plot for Newspaper vs Sales (generated in a previous cell) shows a weak positive correlation. There is a wide spread of data points, suggesting that while there might be a slight tendency for sales to increase with newspaper advertising, the relationship is not very strong.
*   **TV vs Sales:** The scatter plot for TV vs Sales shows a strong positive linear relationship. As TV advertising increases, sales tend to increase significantly. This indicates that TV advertising is likely a strong predictor of sales.
*   **Radio vs Sales:** The scatter plot for Radio vs Sales shows a moderately positive linear relationship. Similar to TV, as radio advertising increases, sales also tend to increase, but the relationship appears to be weaker than that of TV advertising.

The linear regression model you trained (`Linear_Regression` function) attempts to find a linear relationship between the advertising budgets (TV, Radio, Newspaper) and sales. The weights (`w`) learned by the model represent the impact of each advertising channel on sales, and the bias (`b`) represents the baseline sales when all advertising budgets are zero.

Based on the learned weights and bias, the model can predict sales for given advertising budgets. The output of the `predict` function `array([23.87698413])` suggests that with TV=238, Radio=37, and Newspaper=70, the predicted sales are approximately 23.88 units.

To get a more precise interpretation of the model's coefficients, you would typically look at the values of the `w` array. The magnitude of each weight indicates the strength of the relationship, and the sign indicates whether the relationship is positive or negative.
