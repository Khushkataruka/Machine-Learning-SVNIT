```python
import pandas as pd
import numpy as np
```

```python
#x and y must be numpy array
def Linear_Regression(x,y,learning_rate):
  weights=np.zeros((1,x.shape[1]))
  bias=0
  learning_rate=learning_rate
  epochs=500
  for i in range(epochs):
    y_pred=abs(np.dot(weights,x.T)+bias)
    print(y.shape,y_pred.shape)
    loss=np.mean((y-y_pred)**2)
    dw=2*(y-y_pred).dot(x)
    db=2*(y-y_pred)
    weights=weights-learning_rate*dw
    bias=bias-learning_rate*db
    print(f"Loss:{loss},Epoch:{i}\n{'-'*50}\n")
  return weights,bias
```

```python
df=pd.read_csv('Housing.csv')
df.head()
```

```python
df['mainroad']=df['mainroad'].map({'yes':1,'no':0})
df['guestroom']=df['guestroom'].map({'yes':1,'no':0})
df['basement']=df['basement'].map({'yes':1,'no':0})
df['hotwaterheating']=df['hotwaterheating'].map({'yes':1,'no':0})
df['airconditioning']=df['airconditioning'].map({'yes':1,'no':0})
df['prefarea']=df['prefarea'].map({'yes':1,'no':0})
df['furnishingstatus']=df['furnishingstatus'].map({'furnished':1,'semi-furnished':2,'unfurnished':3})
```

```python
# Normalization
# 1. Min max Scaling

area_max=max(df['area'])
area_min=min(df['area'])
price_max=max(df['price'])
price_min=min(df['price'])

df['price']=(df['price']-price_min)/(price_max-price_min)
df['area']=(df['area']-area_min)/(area_max-area_min)

df.head()
```

```python
x=np.array(df.drop('price',axis=1))
y=np.array(df['price']).T
```

```python
w,b=Linear_Regression(x,y,4e-5)
```

```python
std_price=df['price'].std()
mean_price=df['price'].mean()

print(f"Mean of price: {mean_price}")
print(f"Standard deviation of price: {std_price}")
```

```python
def predict(x,w,b):
  return abs(np.dot(w,x.T)+b)
```
